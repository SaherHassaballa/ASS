```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)


url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/amazon_data_webpage.html"
data = requests.get(url).text


soup = BeautifulSoup(data, 'html5lib')

title = soup.title.string
print(f"Title of the page: {title}")


amdata = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"])


amdata2 = {
    "Date": "2021-10-01",
    "Open": 590.10,
    "High": 593.57,
    "Low": 582.58,
    "Close": 590.65,
    "Adj Close": 590.65,
    "Volume": 2577100,
}

new_row = pd.DataFrame([amdata2])

amdata = pd.concat([amdata, new_row], ignore_index=True)

print("Extracted data (method 1):")
print(amdata.head())

amdata = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"])
for row in soup.find("tbody").find_all("tr"):
    col = row.find_all("td")
    date = col[0].text.strip()
    open_price = col[1].text.strip()
    high = col[2].text.strip()
    low = col[3].text.strip()
    close = col[4].text.strip()
    adj_close = col[5].text.strip()
    volume = col[6].text.strip()

    amdata = amdata.append({
        "Date": date,
        "Open": open_price,
        "High": high,
        "Low": low,
        "Close": close,
        "Adj Close": adj_close,
        "Volume": volume
    }, ignore_index=True)

print("Extracted data (method 2):")
print(amdata.head())
```
